#### Task 12: Remove the 'who' variable from the dataframe.

```python
titanic = titanic.drop("who", axis=1)
```

#### Task 13: Fill in the blanks in the 'deck' variable with the most frequent value (mod) of the 'deck' variable.

```python
titanic['deck'].fillna(titanic['deck'].mode().iloc[0], inplace=True)
titanic
```

#### Task 14: Fill in the blank values in the 'age' variable with the median of the 'age' variable.

```python
titanic["age"].fillna(titanic["age"].median(), inplace=True)
titanic
```

#### Task 15: Find the sum, number and mean values of the 'survived' variable at the intersection between the 'pclass' and 'sex' variables.

```python
grup = titanic.groupby(["pclass", "sex"])["survived"]
```

```python
grup.sum()
```

```python
grup.count()
```

```python
grup.mean()
```

```python
grup.agg(["sum", "count", "mean"])
```

#### Task 16: Define the Tips dataset from the Seaborn library.

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

#### Task 17: Find the sum, minimum, maximum and average of the 'total_bill' values according to the categories (Dinner, Lunch) of the 'Time' variable.

```python
tips_grup_time = tips.groupby("time")["total_bill"]
```

```python
tips_grup_time.agg(["sum", "min", "max", "mean"])
```

#### Task 18: Find the sum, minimum, maximum and average of 'total_bill' values by days and times.

```python
tips_day_time = tips.groupby(["day", "time"])["total_bill"]
```

```python
tips_day_time.agg(["sum", "min", "max", "mean"])
```

#### Task 19: Find the sum, minimum, maximum and average of the 'total_bill' and 'type' values of the lunch time and female customers by day.

```python
tips_lunch_female = tips[ (tips["time"] == "Lunch") & (tips["sex"] == "Female")]
```

```python
tips_lunch_female_grup = tips_lunch_female.groupby("day")[["total_bill", "tip"]]
```

```python
tips_lunch_female_grup.agg(["sum", "min", "max", "mean"])
```

#### Task 20: What is the average of orders with 'size' less than 3 and 'total_bill' greater than 10? (use loc).

```python
tip_spec_grup = tips.loc[(tips["size"] < 3) & (tips["total_bill"] > 10)]
```

```python
tip_spec_grup
```

```python
tip_spec_grup["total_bill"].mean()
```
